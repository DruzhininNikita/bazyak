```python
import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """Функция f(x) = sqrt(4x + 7) - 3*cos(x)"""
    return np.sqrt(4*x + 7) - 3*np.cos(x)

print("--- Отделение корней ---")

# ОДЗ: 4x + 7 >= 0 => x >= -1.75. Берем диапазон для поиска
x_vals = np.linspace(-1.75, 5, 500)
y1 = np.sqrt(4 * x_vals + 7)
y2 = 3 * np.cos(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1, label='y = sqrt(4x + 7)', color='blue')
plt.plot(x_vals, y2, label='y = 3*cos(x)', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Отделение корней: sqrt(4x + 7) = 3*cos(x)')
plt.show()

# интервалы, где разность функций меняет знак
for i in range(len(x_vals)-1):
    f_left = math.sqrt(4 * x_vals[i] + 7) - 3 * math.cos(x_vals[i])
    f_right = math.sqrt(4 * x_vals[i+1] + 7) - 3 * math.cos(x_vals[i+1])
    if f_left * f_right < 0:
        print(f"Корень на отрезке [{x_vals[i]:.2f}; {x_vals[i+1]:.2f}]")

print("\n--- Задание 2: Метод итераций ---")

# Параметры для метода итераций
x_curr = 0.5  # начальное приближение из графика
eps = 0.001   # заданная точность
tau = 0.1     # параметр сходимости

print(f"{'n':<4} | {'x_n':<10} | {'f(x_n)':<10} | {'|x_n - x_{n-1}|':<12}")
print("-" * 45)

n = 1
while True:
    # Текущее значение функции
    f_val = math.sqrt(4 * x_curr + 7) - 3 * math.cos(x_curr)

    # Расчет следующего приближения
    x_next = x_curr - tau * f_val
    diff = abs(x_next - x_curr)

    print(f"{n:<4d} | {x_curr:<10.6f} | {f_val:<10.6f} | {diff:<12.6f}")

    if diff < eps:
        break

    x_curr = x_next
    n += 1

print("-" * 45)
print(f"Корень: x = {x_next:.6f}")
print(f"Проверка: f(x) = {math.sqrt(4*x_next + 7) - 3*math.cos(x_next):.6f}")
```

```python
import math

def f(x):
    # Функция f(x) для варианта 6
    return math.sqrt(4 * x + 7) - 3 * math.cos(x)

# Параметры из графического метода
a, b = 0.5, 0.6  # интервал
eps = 0.001      # точность

print(f"{'n':<4} {'a':<10} {'b':<10} {'x':<10} {'f(x)':<10} {'|x_n - x_{n-1}|':<12}")
print('-' * 65)

x_prev = a
n = 1

while True:
    # Формула метода хорд
    x = a - f(a) * (b - a) / (f(b) - f(a))
    diff = abs(x - x_prev)

    print(f"{n:<4d} {a:.6f} {b:.6f} {x:.6f} {f(x):.6f} {diff:.6f}")

    # Условие выхода по точности
    if abs(f(x)) < eps or diff < eps:
        print(f'\nКорень: x = {x:.6f}')
        print(f'Проверка: f({x:.6f}) = {f(x):.6f}')
        break

    # Выбор новой границы интервала
    if f(a) * f(x) < 0:
        b = x
    else:
        a = x

    x_prev = x
    n += 1
```

```python
import math

def f(x):
    """Исходная функция f(x) = sqrt(4x + 7) - 3 * cos(x)"""
    return math.sqrt(4 * x + 7) - 3 * math.cos(x)

# Параметры расчета
x_curr = 0.5  # начальное приближение из графика
eps = 0.001   # заданная точность
tau = 0.1     # коэффициент сходимости (шаг)

print(f"{'n':<4} | {'x_n':<10} | {'f(x_n)':<10} | {'|x_n - x_{n-1}|':<12}")
print("-" * 45)

x_prev = x_curr
n = 1

while True:
    f_val = f(x_curr)

    # Итерационная формула: x_next = x_curr - tau * f(x_curr)
    x_next = x_curr - tau * f_val
    diff = abs(x_next - x_curr)

    print(f"{n:<4d} | {x_curr:<10.6f} | {f_val:<10.6f} | {diff:<12.6f}")

    if diff < eps:
        x_final = x_next
        break

    x_curr = x_next
    n += 1

print("-" * 45)
print(f"Корень: x = {x_final:.4f}")
print(f"Проверка: f({x_final:.4f}) = {f(x_final):.6f}")
```
