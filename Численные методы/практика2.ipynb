```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Приводим к виду f(x) = sqrt(4x + 7) - 3 * cos(x) = 0
def f_np(x):
    return np.sqrt(4*x + 7) - 3*np.cos(x)

# ОДЗ: 4x + 7 >= 0 => x >= -1.75. Берем диапазон от -1.75 до 5.
x_range = np.linspace(-1.75, 5, 1000)
y_range = f_np(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='f(x) = sqrt(4x+7) - 3cos(x)')
plt.axhline(0, color='black', linewidth=1)
plt.grid(True)
plt.title("Локализация корня (Вариант 6)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

# Поиск интервала, где функция меняет знак
print("Поиск интервалов с корнями:")
for i in range(len(x_range)-1):
    if y_range[i] * y_range[i+1] < 0:
        print(f'Корень находится на интервале: [{x_range[i]:.2f}, {x_range[i+1]:.2f}]')

print("-" * 50)

# Используем метод релаксации
def f(x):
    return math.sqrt(4 * x + 7) - 3 * math.cos(x)

x = 0.5  # Начальное приближение на основе графика
eps = 0.001
tau = 0.1  # Параметр шага (чтобы итерации не расходились)

print(f"{'Шаг (n)':<8} | {'x_n':<10} | {'|x_new - x|':<12}")
print("-" * 35)

n = 0
while True:
    f_val = f(x)
    x_new = x - tau * f_val  # Итерационная формула

    diff = abs(x_new - x)
    print(f"{n:<8} | {x:<10.5f} | {diff:<12.5f}")

    if diff < eps:
        x = x_new
        break

    x = x_new
    n += 1

print("-" * 35)
print(f'Найденный корень: {x:.4f}')

# Проверка (подстановка в исходное уравнение)
left = math.sqrt(4 * x + 7)
right = 3 * math.cos(x)
print(f'Проверка: sqrt(4x+7) = {left:.4f}, 3*cos(x) = {right:.4f}')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    """Функция f(x) = sqrt(4x + 7) - 3*cos(x)"""
    return np.sqrt(4*x + 7) - 3*np.cos(x)

print("--- Отделение корней ---")

# ОДЗ: 4x + 7 >= 0 => x >= -1.75
x_vals = np.linspace(-1.75, 4, 400)
y1 = np.sqrt(4 * x_vals + 7)
y2 = 3 * np.cos(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1, label='y = sqrt(4x + 7)', color='blue', linewidth=1.8)
plt.plot(x_vals, y2, label='y = 3*cos(x)', color='red', linewidth=1.8)
plt.grid(True)
plt.legend()
plt.title('Отделение корней уравнения sqrt(4x + 7) = 3*cos(x)')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Поиск интервалов смены знака разности функций
for i in range(len(x_vals)-1):
    f_left = math.sqrt(4 * x_vals[i] + 7) - 3 * math.cos(x_vals[i])
    f_right = math.sqrt(4 * x_vals[i+1] + 7) - 3 * math.cos(x_vals[i+1])
    if f_left * f_right < 0:
        print(f"Корень находится на отрезке [{x_vals[i]:.2f}; {x_vals[i+1]:.2f}]")

print("\n--- Задание 2: Метод итераций ---")

x_curr = 0.5  # Начальное приближение из графика
eps = 0.001   # Требуемая точность
tau = 0.1     # Коэффициент сходимости

print(f"{'n':<4} | {'x_n':<10} | {'f(x_n)':<10} | {'|x_next - x_n|':<12}")
print("-" * 45)

n = 1
while True:
    # Вычисляем значение функции f(x)
    f_val = math.sqrt(4 * x_curr + 7) - 3 * math.cos(x_curr)

    x_next = x_curr - tau * f_val
    diff = abs(x_next - x_curr)

    print(f"{n:<4d} | {x_curr:<10.6f} | {f_val:<10.6f} | {diff:<12.6f}")

    if diff < eps:
        break

    x_curr = x_next
    n += 1

print("-" * 45)
print(f"Корень: x = {x_next:.6f}")
print(f"Проверка: f({x_next:.6f}) = {math.sqrt(4*x_next + 7) - 3*math.cos(x_next):.6f}")
```
