Задание 1

```python
def iteration_step(val):
    """Итерационная функция phi(x) = -0.5x + 1.5"""
    return -0.5 * val + 1.5

domain = np.linspace(0, 2, 200)
start_point = 0.2
steps = 8

plt.figure(figsize=(8, 8))

plt.plot(domain, iteration_step(domain), color='red', label=r'$\phi(x) = -0.5x + 1.5$')
plt.plot(domain, domain, color='green', linestyle=':', label='Линия $y=x$')

current_x = start_point
history_x, history_y = [], []

for _ in range(steps):
    next_x = iteration_step(current_x)
    
    plt.plot([current_x, current_x], [current_x, next_x], 'orange', lw=1)
    plt.plot([current_x, next_x], [next_x, next_x], 'orange', lw=1)
    plt.scatter(current_x, next_x, c='black', s=25, zorder=5)
    
    current_x = next_x

plt.axhline(0, color='black', lw=0)
plt.axvline(0, color='black', lw=1)
plt.title('Графическая интерпретация сходящейся итерации')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='upper right')
plt.grid(which='both', linestyle='--')
plt.xlim(0, 2)
plt.ylim(0, 2)

plt.show()
```

### Задание 2

```python
def f(x):
    return x * np.sin(x) - 1

curr_x = 1.1
step_weight = 0.4
eps = 0.00001

print("Решение уравнения методом итераций:")
print("-" * 30)

for i in range(1, 21):
    next_x = curr_x - step_weight * f(curr_x)
    
    print(f"Шаг {i}: x = {next_x:.7f}")
    
    if abs(next_x - curr_x) < eps:
        print("-" * 30)
        print(f"Корень найден: {next_x:.6f}")
        print(f"Проверка (f(x)): {next_x * np.sin(next_x) - 1:.8f}")
        break
        
    curr_x = next_x
```
