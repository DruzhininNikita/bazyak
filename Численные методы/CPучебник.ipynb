```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Задание 1

```python
def f(x):
    """функция f(x): sqrt(4x + 7) - 3*cos(x)"""
    return np.sqrt(4 * x + 7) - 3 * np.cos(x)


print("отделение корней на отрезке [-1, 1]")

x = np.linspace(-1, 1, 400)

y1 = np.sqrt(4 * x + 7)
y2 = 3 * np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='sqrt(4x + 7)', color='blue', linewidth=1.8)
plt.plot(x, y2, label='3cos(x)', color='red', linewidth=1.8)
plt.grid(True)
plt.legend()
plt.title('отделение корней уравнения sqrt(4x + 7) = 3cos(x)')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1, 1)
plt.show()

for i in range(len(x)-1):
    if (y1[i] - y2[i]) * (y1[i+1] - y2[i+1]) < 0:
        print(f"корень на отрезке [{x[i]:.2f}; {x[i+1]:.2f}]")
```

```python
def f(x):
    """функция f(x): sqrt(4x + 7) - 3*cos(x)"""
    return np.sqrt(4 * x + 7) - 3 * np.cos(x)
print("метод половинного деления")

a, b, eps, n = 0.35, 0.4, 0.00001, 0
print(f"начальный отрезок: [{a}, {b}]")

while (b - a) > eps:
    c = (a + b) / 2
    if f(a) * f(c) <= 0:
        b = c
    else:
        a = c
    n += 1

print(f"корень:  x = {(a + b) / 2:.6f}")
print(f"количество итераций: {n}")
```

Задание 3

```python
def f(x):
    """функция f(x): sqrt(4x + 7) - 3*cos(x)"""
    return np.sqrt(4 * x + 7) - 3 * np.cos(x)

def phi(x):
    lam = -0.1
    return x + lam * f(x)

print("метод итераций для корня на [0.35; 0.4]")
x0 = 0.35
eps = 0.00001
x = x0
n = 0

while True:
    x_new = phi(x)

    if abs(x_new - x) < eps:
        x = x_new
        break
    x = x_new
    n += 1

print(f"Корень: x = {x:.6f}")
print(f"Количество итераций: {n}")
```
