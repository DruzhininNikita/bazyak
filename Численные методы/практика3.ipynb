```python
import math

x = 0.5  # начальное значение (близко к пересечению графиков)
eps = 0.001
tau = 0.1  # параметр шага для обеспечения сходимости (метод релаксации)

print(f"{'n':<4} | {'x_n':<10} | {'|x_new - x|':<12}")
print("-" * 32)

n = 0
while True:
    # f(x) - это разница между левой и правой частью
    f_x = math.sqrt(4 * x + 7) - 3 * math.cos(x)

    # Итерационная формула: x_{n+1} = x_n - tau * f(x_n)
    x_new = x - tau * f_x

    diff = abs(x_new - x)
    print(f"{n:<4} | {x:<10.5f} | {diff:<12.5f}")

    if diff < eps:
        x = x_new
        break

    x = x_new
    n += 1

print("-" * 32)
print(f'Корень: {x:.4f}')
# Проверка: подставляем корень в исходное уравнение
left_side = math.sqrt(4 * x + 7)
right_side = 3 * math.cos(x)
print(f'Проверка: {left_side:.4f} = {right_side:.4f}')
```
